Данные студентов представлены в виде списка списков:

Преобразуйте данные таким образом, чтобы они были представлены в виде словаря, в котором ключами являются номера студентов, а значениями — список из других данных о студентах.

Напишите функцию, которая принимает в качестве аргумента номер студента и новый номер группы и позволяет изменить номер группы этого студента на новый внутри созданного словаря.

Напишите функцию, которая принимает номер группы и выводит список (фамилии, имена, отчества) всех студентов этой группы. Необходимо обращаться к словарю.

```python
students = [
["0001", "Антонов", "Антон", "Игоревич", "20.08.2009","БСТ161"],
["1102", "Богов", "Артем", "Игоревич", "25.01.2010","БСТ162"],
["0333", "Глаголева", "Анастасия", "Николаевна", "11.07.2009", "БСТ163"],
["4004", "Степанова", "Наталья", "Александровна", "13.02.2008", "БСТ161"],
["0045", "Боков", "Игорь", "Харитонович", "02.06.2009", "БСТ161"],
["0096", "Васильков", "Валентин", "Сергеевич", "20.03.2009", "БСТ164"],
["0607", "Сиропова", "Виолетта", "Эдуардовна", "28.05.2010", "БСТ162"]
    ]
```

```python
# преобразуем словарь (ключ - номер студента, значение - список данных)
students_dict = {}
for i in range(0, len(students)):
    students_dict[students[i][0]]=students[i][1:]
print (students_dict)
```

```python
# по номеру студента меняем его группу
def nomer_group(number, new_group):
    global students_dict
    if number in students_dict:
        students_dict[number][-1] = new_group
nomer_group('0096', 'БСТ163')
print(students_dict)
# у студента Василькова (предпоследний в словаре) поменялась группа с БСТ164 на БСТ163
```

```python
# выводим фио студентов по номеру группы
def stud_group(group):
    global students_dict
    for id, value in students_dict.items():
        if(value[-1] == group):
            for element in value[0:3]:
                print(element, end=' ')
            print()
stud_group('БСТ163')
```
