```python
students_dict = {}
for i in range(0, len(students)):
    students_dict[students[i][0]]=students[i][1:]
    print (students_dict)
```

```python
def n_group(number,new):
    students_dict[number][4] = new
n_group('0096','БСТ163')
```

```python
def spisok(group):
    for key,values in students_dict.items():
        if values[4] == group:
            print(values[0],values[1],values[2])
spisok('БСТ163')
```
